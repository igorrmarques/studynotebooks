```python
import pandas as pd

url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep= ';')
dados.head()
```

```python
dados = dados.fillna(0)
dados['Valor_Mensal'] = dados['Valor'] + dados['Condominio']
dados['Valor_Anual'] = dados['Valor_Mensal']*12 + dados['IPTU']
dados.head()
```

```python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()
```

```python
dados['Possui_Suite'] = dados['Suites'].apply(lambda x: 'Possui' if x > 0 else 'Não possui')
dados.head()
```

```python
dados.to_csv('../dataprocessed/dados_completos.csv', index= False, sep='-')
```
